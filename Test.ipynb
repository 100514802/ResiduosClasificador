```python
import os
import torch
from torchvision.models import resnet50

device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

# Define el modelo con la misma arquitectura que usaste para entrenar
model = resnet50(pretrained=False, num_classes=6)  # Ajusta 'num_classes' según tu configuración

# Especifica la ruta donde guardaste el modelo entrenado
models_path = os.path.join('..','PIC', 'models')

# Carga los parámetros entrenados en el modelo
model = torch.load(os.path.join(models_path, 'ResNet50.pt')).to(device)
model.eval()  # Cambia el modo del modelo a evaluación
```

```python
from torchvision import transforms
from PIL import Image
import os

# Especifica la carpeta donde se encuentran las imágenes de prueba
carpeta_prueba = 'C:/Users/Quique/OneDrive - Universidad Carlos III de Madrid/Master Robotica/Primer Semestre/Procesamiento de imagenes por computador/PIC/dataset-test'

# Lista todas las imágenes en la carpeta de prueba
imagenes_prueba = [f for f in os.listdir(carpeta_prueba) if os.path.isfile(os.path.join(carpeta_prueba, f))]

# Para cada imagen en la carpeta de prueba
for imagen in imagenes_prueba:
    # Construye la ruta completa de la imagen
    ruta_imagen = os.path.join(carpeta_prueba, imagen)

    # Carga y preprocesa la imagen de prueba
    img = Image.open(ruta_imagen)
    preprocess = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    img_tensor = preprocess(img)
    img_tensor = img_tensor.unsqueeze(0)  # Agrega dimensión de lote (batch)

    # Realiza la predicción
    with torch.no_grad():
        output = model(img_tensor)

    # Obtén la etiqueta predicha
    _, predicted_idx = torch.max(output, 1)
    predicted_label = predicted_idx.item()

    # Carga las etiquetas del conjunto de datos (ajusta según tus etiquetas)
    class_labels = ["clase1", "clase2", "clase3", "clase4", "clase5", "clase6"]

    # Muestra la predicción para cada imagen en la carpeta de prueba
    print(f'Imagen: {imagen}, Predicción: {class_labels[predicted_label]}')
```
